## Building user profiles

For each combination of user and topic, we compute the dot products over all the questions between the user feedback and if the question is related to the topic. 

We are basically aggregating by counting for each user and topic, how many times he liked (adding) or disliked (substracting) the topics contained in the questions he provided feedback for.

```python
def buildUserProfiles(df):
    profiles = {}
    userIndex = []
    for user in users.columns:
        userIndex.append(user)
        for topic in df.columns:
            if topic not in profiles:
                profiles[topic] = [];
            profiles[topic].append(users[user].dot(df[topic]));
    profiles = pd.DataFrame(profiles, index=userIndex)[df.columns.tolist()]
    return profiles

buildUserProfiles(df)
```

## Predicting user preferences

Based on the user profiles that we just built, we'll evaluate the similarity between that profile and each question by applying the dot product between them.

```python
def dot2(a, b):
    result = 0.0;
    for n in np.arange(len(a)):
        result += a[n] * b[n]
    return result

def computePredictions(df, profiles):
    predictions = {}
    index = []

    def usersToItemDistance(item):
        def userToItemDistance(user):
            if user.name not in predictions:
                predictions[user.name] = [];
            #predictions[user.name].append(user.dot(item));
            predictions[user.name].append(dot2(user, item));
        profiles.apply(userToItemDistance, axis=1);
    df.apply(usersToItemDistance, axis=1);
    predictions = pd.DataFrame(predictions)
    return predictions

computePredictions(df, buildUserProfiles(df))
```

Based on those scores, we can count how many questions each user would like, dislike or be neutral to

```python
def displayPreferences(predictions):
    print "TOTAL"
    likes = "Likes\t"
    for user in users.columns:
        likes += "\t" + str(len(predictions[predictions[user] > 0]))
    print likes
    dislikes = "Dislikes"
    for user in users.columns:
        dislikes += "\t" + str(len(predictions[predictions[user] < 0]))
    print dislikes
    neutral = "Neutral\t"
    for user in users.columns:
        neutral += "\t" + str(len(predictions[predictions[user] == 0]))
    print neutral

displayPreferences(computePredictions(df, buildUserProfiles(df)))
```

Based on those predictions, the top 5 questions for each user would be:

```python
def extractTop(predictions, N):
    topN = {}
    for user in users.columns:
        top = predictions.sort(user, ascending=False)[user]
        topN[user] = top[:N].index
        #print top[:N]
    return pd.DataFrame(topN)

extractTop(computePredictions(df, buildUserProfiles(df)), 5)
```

# IDF

Instructions:

<i>With the unit weight applied, now evaluate the topics relevance using IDF. The higher
the number of questions a topic has, the lower its relevance is. Rare topics would have more
weight applying IDF now, thus being more relevant for the final prediction. With the new
predictions, provide the top-5 questions recommended per each user.</i>

```python
def computeIDF(df):
    return 1 / df.apply(sum)
    
computeIDF(df_normalized)
```

```python
def dot3(a, b, c):
    result = 0.0;
    for n in np.arange(len(a)):
        result += a[n] * b[n] * c[n]
    return result

def computePredictionsIDF(df, profiles):
    predictions = {}
    index = []
    idf = computeIDF(df)
    
    def usersToItemDistance(item):
        def userToItemDistance(user):
            if user.name not in predictions:
                predictions[user.name] = [];
            predictions[user.name].append(dot3(user, item, idf));
        profiles.apply(userToItemDistance, axis=1);
    df.apply(usersToItemDistance, axis=1);
    predictions = pd.DataFrame(predictions)
    return predictions

computePredictionsIDF(df_normalized, buildUserProfiles(df_normalized))
```

# Tweaking the given instructions

Since the instructions of the assignment is guiding us towards using the distance between profiles and items by performing a dot product. I wanted to check if we would get different results if we were using the proper definition of cosine distance instead.

The only difference is that we would need to divide the dot product by the norm of the vectors as they are not of unit length.

```python
def computePredictionsIDFcorrected(df, profiles):
    predictions = {}
    index = []
    idf = computeIDF(df)
    
    def usersToItemDistance(item):
        def userToItemDistance(user):
            if user.name not in predictions:
                predictions[user.name] = [];
            #predictions[user.name].append(dot3(user, item, idf));
            predictions[user.name].append(dot3(user, item, idf) / (np.linalg.norm(user) * np.linalg.norm(item) * np.linalg.norm(idf)));
        profiles.apply(userToItemDistance, axis=1);
    df.apply(usersToItemDistance, axis=1);
    predictions = pd.DataFrame(predictions)
    return predictions

computePredictionsIDFcorrected(df_normalized, buildUserProfiles(df_normalized))
```

```python
displayPreferences(computePredictionsIDFcorrected(df_normalized, buildUserProfiles(df_normalized)))
```

```python
extractTop(computePredictionsIDFcorrected(df_normalized, buildUserProfiles(df_normalized)), 5)
```

As we can notice here, using the proper definition of the cosine distance or the simple dot product calculation is not affecting the recommendations that we are getting in the end for this particular dataset.

## Building user profiles (Content-based with Average user profile)

Another solution to the cold start problem is to build an average user profile over all the users. When we encounter a new user that we don't know anything about yet, we can use the average user profile instead and continue to provide recommendations based on content for this unknown user.

This method would be equivalent to fill the unknown user's ratings with an average of all the other users for each question.

This approach would take into account the popular topics that our users generally prefer reading about (in a non personalized sense) and give recommendations on the questions including such topics. This is a slight improvement over the previous non-personalized method because we will be able to recommend new questions that didn't became so popular yet but that includes popular topics instead. The previous method will on the other hand always recommend the most popular questions (and because of its recommendations, these questions may remain the top populars ones) which may not change much over time.

```python
def isEmptyArray(v):
    for n in v:
        if n != 0:
            return False
    return True;

def fixColdStart(profiles):
    coldStart = profiles[profiles.apply(isEmptyArray, axis=1)]
    notColdStart = profiles.ix[profiles.index - coldStart.index]
    avgProfile = notColdStart.apply(np.mean)
    def fillEmptyArray(row):
        if isEmptyArray(row):
            return avgProfile
        else:
            return row
    return profiles.apply(fillEmptyArray, axis=1)

fixColdStart(buildUserProfiles(df_normalized))
```

## Building Collaborative Filtering recommendations

Let's reuse the Collaborative Filtering recommender from the previous assignment in order to produce recommendations on questions for each user instead of movies.

When the user base and question base will grow, the collaborative filtering approach will be much more successfull at distinguishing similarities between users and their preferences in terms of questions.

```python
from math import sqrt

def getCommonRatings(prefs, person1, person2):
  x = [];
  y = [];
  for i in np.arange(len(prefs)):
    if prefs[person1].ix[i] != 0 and prefs[person2].ix[i] != 0:
      x.append(prefs[person1][i]);
      y.append(prefs[person2][i]);
  return x, y;

def average(x):
    if len(x) > 0:
      return float(sum(x)) / len(x)
    else:
      return 0

def pearsonSimilarity(prefs, person1, person2):
  r=0.0
  x,y = getCommonRatings(prefs, person1, person2);
  # Following code is actually 
  # from http://stackoverflow.com/questions/3949226/calculating-pearson-correlation-and-significance-in-python
  n = len(x)
  avg_x = average(x)
  avg_y = average(y)
  diffprod = 0
  xdiff2 = 0
  ydiff2 = 0
  for idx in range(n):
      xdiff = x[idx] - avg_x
      ydiff = y[idx] - avg_y
      diffprod += xdiff * ydiff
      xdiff2 += xdiff * xdiff
      ydiff2 += ydiff * ydiff
  if (xdiff2 * ydiff2) > 0:
    r = diffprod / sqrt(xdiff2 * ydiff2)
  else:
    r = 0.0;
  return r

def getRanking(prefs, person, questionIndex, similarity=pearsonSimilarity):
    totalWeights = 0.0;
    rating = 0.0;
    for key in prefs:
        if person != key and prefs[key].ix[questionIndex] != 0:
          weight = pearsonSimilarity(prefs, person, key);
          rating += weight * (prefs[key].ix[questionIndex]);
          totalWeights += abs(weight);
    if totalWeights != 0:
        return (rating / totalWeights, questionIndex);
    else:
        return (0.0, questionIndex);

def getRecommendations(prefs, person, similarity=pearsonSimilarity):
    rankings = []
    for key in np.arange(len(df)):
        if prefs[person].ix[key] == 0:
            rankings.append(getRanking(prefs, person, key, similarity));
    rankings = sorted(rankings, reverse=True);
    return rankings

def extractTopCollaborative(prefs, N):
    topN = {}
    for user in prefs.columns:
        if user in activeUsers:
            top = getRecommendations(prefs, user)
            for i in np.arange(N):
                if (not user in topN):
                    topN[user] = [];
                topN[user].append(top[i][1])
        else:
            topN[user] = avgRating[:N].index
    return pd.DataFrame(topN)

extractTopCollaborative(users, 5)
```

## Building Popularity aware Collaborative Filtering recommendations

But now that we integrated a Collaborative Filtering method, we can exploit the data from the up votes and down votes on each user answers. We can indeed build a popularity profile for each user and each topic as we did previously with the content based recommendations. We can therefore model how knowledgeable or relevant the answers of certain user are towards a topic, judged by the community and use this measure when weighing the ratings of the user.

Let's first build for each user, their popularity profiles (how popular their answers are judged by the community to each topic):
This would be a kind of trust that the community is having into the answers of an individual

```python
def buildUserTrustProfiles(df):
    profiles = {}
    userIndex = []
    for user in answers.columns:
        userIndex.append(user)
        for topic in df.columns:
            if topic not in profiles:
                profiles[topic] = [];
            profiles[topic].append(answers[user].dot(df[topic]));
    profiles = pd.DataFrame(profiles, index=userIndex)[df.columns.tolist()]
    return profiles

trustProfiles = buildUserTrustProfiles(df_normalized)
trustProfiles
```

Next step is to re-use the content based approach to build predictions of how trustful each user is towards each question if he were to answer it. This is done by using the same method as content-based approach with the tf-IDF normalization but using the popularity profiles instead of the user preferences profiles:

```python
trustPredictions = computePredictionsIDF(df_normalized, buildUserTrustProfiles(df_normalized))
trustPredictions
```

Now, let's integrate this new data into the Collaborative Filtering approach: when computing the weight of other users to predict someone's preferences towards a new question, we also include the popularity score of the other users so that knowledgeable people in certain topics have higher influence than someone that is judged less interesting by the community. 

```python
def getTrustRanking(prefs, person, questionIndex, similarity=pearsonSimilarity):
    totalWeights = 0.0;
    rating = 0.0;
    for key in prefs:
        if person != key and prefs[key].ix[questionIndex] != 0:
            weight = pearsonSimilarity(prefs, person, key) * trustPredictions[key].ix[questionIndex];
            rating += weight * (prefs[key].ix[questionIndex]);
            totalWeights += abs(weight);
    if totalWeights != 0:
        return (rating / totalWeights, questionIndex);
    else:
        return (0.0, questionIndex);
    

def getTrustRecommendations(prefs, person, similarity=pearsonSimilarity):
    rankings = []
    for key in np.arange(len(df)):
        if prefs[person].ix[key] == 0:
            rankings.append(getTrustRanking(prefs, person, key, similarity));
    rankings = sorted(rankings, reverse=True);
    return rankings

def extractTopTrustCollaborative(prefs, N):
    topN = {}
    for user in prefs.columns:
        if user in activeUsers:
            top = getTrustRecommendations(prefs, user)
            for i in np.arange(N):
                if (not user in topN):
                    topN[user] = [];
                topN[user].append(top[i][1])
        else:
            topN[user] = avgRating[:N].index
    return pd.DataFrame(topN)

extractTopTrustCollaborative(users, 5)
```

## Stacking recommendations

As we have noticed, the results from the Collaborative Filtering approach are not very good on this dataset. It is doutbul that including the trust measure would improve such results as well. However, it is most likely to be a very good method when the size of user base and question base would be much larger.

As the number of users and questions of such website is certainly going to grow in numbers, the Collaborative Filtering and Trust aware Collaborative Filtering approaches will gain accuracy. As a result, it would be important to design a stacking method that would decide to use either the content based recommendations, the collaborative filtering recommendations or the trust-aware collaborative filtering recommendations.

- Therefore, for brand new users, the non-personalized approach using either the most popular questions or the content-based average profile could be activated.
- When such users start using and rating more and more questions, we will be able to understand their preferences a little more and start building their content-based preference profile to activate content based recommendations.
- After some time, the user would have given enough data so we are able to match his behavior to other similar users and exploit results from the Collaborative Filtering approach to produce better recommendations. We could even include at this point the popularity aware collaborative approach too that would take into account feedback from the community on the relevancy of each users writing an answer.

Unfortunately, since we are lacking further data, we wouldn't be able to evaluate at this point and learn the best parameters and thresholds to switch between the difference recommendation methods.

## Combining recommendations

Considering how a website like quora would work. The recommendations may not be limited to a small number of questions.

When a user scrolls down through the top recommendations and start going deeper and deeper, it might signify that this user is looking for more interesting questions than just going through the list of past recommendations.

Therefore, on such particular behavior, it might be beneficial to switch the recommendations to a combination strategy where new questions outside of the user's bubble could be thrown in:
Top 10 questions may follow the standard strategy but down the list, we may want to introduce new content that is not based from the behavior of similar users or content of previously liked questions. We could suggest a random recommendation with a higher probability when scrolling down through the list of questions. 

Or we could even use some reinforcement learning policies that would try to explore the user's preferences on topics that he never read before in order to understand his tastes towards them. The system could then learn more details about his profile and in the long term suggest more relevant questions on a more diverse range of topics.

Of course, we would need to be careful and mix in with a reasonable proportion of regular recommendations or the user may think that the recommendations are not working at all past a certain point down the list of questions.
