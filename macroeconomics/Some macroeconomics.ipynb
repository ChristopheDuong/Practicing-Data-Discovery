# Merge csv files into common data structure

```python
rawdata = {};
allKeys = set();

def toDate(d):
    return datetime.datetime.strptime (d, "%d/%m/%Y %H:%M:%S %p");

def extractYear(d):
    return (toDate(d).year);

def extractData(row):
    print row
    
for files in os.listdir("."):
    if files[-4:] == ".csv":
        for key1, data1 in pd.read_csv(files).groupby('country'):
            key1 = key1.lower();
            data1['year'] = data1['Date'].apply(extractYear);
            if key1 in rawdata:
                countrydata = rawdata[key1];
            else:
                countrydata = {};
            for key2, data2 in data1.groupby('year'):
                if key2 in countrydata:
                    yearData = countrydata[key2];
                else:
                    yearData = {};
                if 'indicator' in data2.keys():
                    keys = ['indicator', 'Value'];
                else:
                    keys = ['subject', 'Value'];
                for element in data2[keys].values:
                    allKeys.add(element[0]);
                    yearData[element[0]] = element[1];
                countrydata[key2] = yearData;
            rawdata[key1] = countrydata;
```

# Convert to Dataframe

```python
data = {};

def addData(data, key, value):
    if key not in data:
        data[key] = [];
    data[key].append(value);
    return data;

for i in rawdata:
    # if we have all dates between 2005-2012
    if len(rawdata[i]) == 8:
        for j in rawdata[i]:
            insertedKey = set();
            addData(data, 'country', i);
            addData(data, 'year', j);
            for k in rawdata[i][j]:
                insertedKey.add(k);
                addData(data, k, rawdata[i][j][k]);
            for key in allKeys:
                if key not in insertedKey:
                    addData(data, key, None);
df = pd.DataFrame(data);
# display 5 randomly selected rows
df.loc[np.random.choice(df.index, 5, replace=False)]
```

# Reconcile countries entered under different names

```python
from difflib import SequenceMatcher

df = pd.DataFrame(data);

def similar(a, b):
    return SequenceMatcher(None, a, b).ratio()

def mergeData(df, str1, str2):
    print "Merging data from '" + str1 + "' into '" + str2 + "'";
    df1 = df[df['country'] == str1];
    df2 = df[df['country'] == str2];
    for key in df.keys():
        if len(df1[df1[key].notnull()]) > 0 and len(df2[df2[key].notnull()]) == 0:
            df.ix[df2.index, key] = df1[key].values;
    return df.drop(df1.index);        

manualMatched = {
    "irak" : "iraq",
    "united states" : "united states of america",
    "bosnia and herzegovina" : "bosnia",
    "taiwan province of china" : "taiwan",
    "islamic republic of iran" : "iran",
    "côte d'ivoire" : "ivory coast",
    "slovak republic" : "slovakia",
    "lao p.d.r." : "laos",
    "trinidad and tobago" : "trinidad",
    "brunei darussalam" : "brunei"
};

missingCarSales = set();
for key, value in df.groupby('country'):
    if len(value[value['All vehicles'].isnull()]) > 0 and key not in manualMatched:
        missingCarSales.add(key);
   
missingMacroEconomics = set();
for key, value in df.groupby('country'):
    if len(value[value['Gross domestic product, current prices (U.S. dollars)'].isnull()]) > 0 and key not in manualMatched:
        missingMacroEconomics.add(key);

ignoredMatched = {
    "haiti" : "tahiti",
    "iraq" : "iran",
    "kosovo" : "kosovo"
};

for key, value in manualMatched.iteritems():
    df = mergeData(df, key, value);
    
for str1 in missingCarSales:
    for str2 in missingMacroEconomics:
        if str1 in ignoredMatched.keys() and ignoredMatched[str1] == str2:
            continue;
        if similar(str1, str2) > 0.7:
            df = mergeData(df, str2, str1);
for key, value in df.groupby('country'):
    if len(value[value['All vehicles'].isnull()]) > 0 and key not in manualMatched:
        print("Dropping data from '" + key + "'");
        df = df.drop(df[df['country'] == key].index);
for key, value in df.groupby('country'):
    if len(value[value['Gross domestic product, current prices (U.S. dollars)'].isnull()]) > 0 and key not in manualMatched:
        print("Dropping data from '" + key + "'");
        df = df.drop(df[df['country'] == key].index);
```

# Build a second dataframe of correlations against a target variable

This second dataframe is built around a target variable that we should select.

As an example, let's pick 'Passenger cars' which represents the value of passenger cars sold.

We then compute the correlation of this variable with all the other ones, using different groupings: "Overall' (no groupings), or by country or by year. (the type of groupings is declared in the column 'type')

Then we can compute an average in each of these subgroups, the 'type' column is set to 'Average' and the key is the subgroup on which the average was computed.

```python
#targetVariable = 'All vehicles';
targetVariable = 'Passengers Cars';
dependencies = {};

def computeCorrelation(df, targetVariable, variable):
    filteredDf = df[(df[targetVariable].notnull() & df[variable].notnull())];
    if len(filteredDf) > 0:
        df1 = filteredDf[targetVariable];
        df2 = filteredDf[variable];
        p = pearsonr(df1, df2);
        return p[0];
    else:
        return None;
    
addData(dependencies, 'key', 'Overall');
addData(dependencies, 'type', 'Global');
for i in df.keys():
    if i != targetVariable and i != 'country' and i != 'year':
        addData(dependencies, i, computeCorrelation(df, targetVariable, i));
for key, value in df.groupby('country'):
    addData(dependencies, 'key', key);
    addData(dependencies, 'type', 'country');
    for i in df.keys():
        if i != targetVariable and i != 'country' and i != 'year':
            addData(dependencies, i, computeCorrelation(value, targetVariable, i));
for key, value in df.groupby('year'):
    addData(dependencies, 'key', str(key));
    addData(dependencies, 'type', 'year');
    for i in df.keys():
        if i != targetVariable and i != 'country' and i != 'year':
            addData(dependencies, i, computeCorrelation(value, targetVariable, i));          
for key, value in pd.DataFrame(dependencies, dependencies['key']).groupby('type'):
    addData(dependencies, 'key', key);
    addData(dependencies, 'type', 'Average');
    for i in df.keys():
        if i != targetVariable and i != 'country' and i != 'year':
            addData(dependencies, i, value[i].mean());
corr = pd.DataFrame(dependencies, dependencies['key'])
corr.loc[np.random.choice(corr.index, 5, replace=False)]            
```

Considering all the values from each columns, we can look at the 'overall' correlation:
