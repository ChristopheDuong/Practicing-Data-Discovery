# Using python-LinkedIn library to retrieve data from API

```python
from linkedin import linkedin # pip install python-linkedin
import json

# Instantiate the developer authentication class
auth = linkedin.LinkedInDeveloperAuthentication(CONSUMER_KEY, CONSUMER_SECRET, 
                                USER_TOKEN, USER_SECRET, 
                                RETURN_URL, 
                                permissions=linkedin.PERMISSIONS.enums.values())

# Pass it in to the app...
app = linkedin.LinkedInApplication(auth)
```

# Download all group memberships of current user

```python
groups = app.get_memberships(selectors=['group:(id,name,description,num-members)'],
                             params={ 'count' : 35 });
f = open("groups.json", 'w')
f.write(json.dumps(groups, indent=1))
f.close()
```

### Extract group ids

```python
ids = [];
group_ids = [];
group_name = {};

for group in groups["values"]:
    key = int(group["_key"]);
    ids.append((int(group["group"]["numMembers"]), key));
    group_name[key] = group['group']['name'];
ids.sort(reverse=True);
for i in ids:
    if i[1] not in [ 39169, 1982164, 41001, 2212808, 4270572, 42939, 86782 ]:
        group_ids.append(i[1]);
```

# Download all posts and comments from each group

```python
for i in group_ids:
    print "From group:" + group_name[i];
    retrievedPosts = -1;
    start = 0;
    values = [];
    while (retrievedPosts != 0):
        posts = app.get_posts(i,
                                selectors = [
                                    'id,creation-timestamp,title,summary,site-group-post-url,creator:(id,first-name,last-name,headline),comments:(text,creation-timestamp,creator:(id,first-name,last-name,headline))'
                                ],
                                params = {
                                    'category' : 'discussion',
                                    'order' : 'recency',
                                    'count' : 100,
                                    'start' : start
                                });
        if 'values' in posts:
            values += posts['values'];
            retrievedPosts = len(posts['values']);
            start += retrievedPosts;
            print(retrievedPosts);
        else:
            break;
    posts['values'] = values;
    allPosts[i] = posts;

f = open("posts.json", 'w');
f.write(json.dumps(allPosts, indent=1));
f.close();
```

# Extract common nouns of each group

```python
import  nltk 
import matplotlib.pyplot as plt
from wordcloud import WordCloud
import os

%matplotlib inline

if not os.path.exists("group_wordcloud"):
    os.makedirs("group_wordcloud"); 
c = 0;    
for group in allPosts:
    groupPosts = allPosts[group];
    if 'values' in groupPosts:
        groupTokens = [];
        groupText = "";
        for post in groupPosts['values']:
            if ('title' in post):
                text = post['title'];
            if ('summary' in post):
                text += " " + post['summary'];    
            if ('comments' in post and post['comments']['_total'] > 0):
                for comment in post['comments']['values']:
                    text += " " + comment['text'];
            groupText += " " + text.lower();
            tokens = nltk.word_tokenize(text.lower());
            groupTokens = groupTokens + nltk.pos_tag(tokens);
        word_tag_fd = nltk.FreqDist(groupTokens);
        if c < 5:
            print ""
            print ""
            print "In group: " + group_name[int (group)];
            print ""
            print ""
        if len(groupText) > 10 and c < 5:
            wordcloud = WordCloud().generate(groupText)
            plt.imshow(wordcloud)
            plt.axis("off")
            plt.savefig('group_wordcloud/' + group_name[int (group)] + '.png')
            plt.show()
            c += 1;
        groupsCounts = {};
        for (wt, count) in word_tag_fd.most_common():
            if wt[1].startswith("NN") and count > 10:
                groupsCounts[wt[0]] = count;
                #print wt[0], count
        postCounts[group_name[int (group)]] = groupsCounts;

f = open("postCounts.json", 'w');
f.write(json.dumps(postCounts, indent=1));
f.close();
```

# Extract Users participating in each group

```python
import pandas as pd
import numpy as np

users = {};
  
def extractInfo(obj, field):
    if field in obj:
        return obj[field];
    else:
        return None;
    
def updateUser(post, activity):
    user_id = extractInfo(post['creator'], 'id');
    if user_id in users:
        user = users[user_id];
    else:
        user = {};
    user['id'] = user_id;
    user['firstName'] = extractInfo(post['creator'], 'firstName');
    user['lastName'] = extractInfo(post['creator'], 'lastName');
    user['headline'] = extractInfo(post['creator'], 'headline');
    s = activity;
    if s not in user:
        user[s] = 0;
    user[s] += 1;
    users[user_id] = user;

for groupId in allPosts:
    groupPosts = allPosts[groupId];
    groupId = int(groupId);
    if 'values' in groupPosts:
       for post in groupPosts['values']:
            if ('creator' in post):
                updateUser(post, 'discussion in ' + group_name[groupId]);
            if ('comments' in post and post['comments']['_total'] > 0):
                for comment in post['comments']['values']:
                   updateUser(comment, 'comment in ' + group_name[groupId]);
data = {
        'id' : [],
        'firstName' : [],
        'lastName' : [],
        'headline' : []
};
activity_group = [];

for key, groupName in group_name.items():
    for activity in ['discussion in ', 'comment in ']:
        data[activity + groupName] = [];
        activity_group.append(activity + groupName);

# Massage data into dataframes
for user in users.values():
    data['id'].append(user['id']);
    data['firstName'].append(user['firstName']);
    data['lastName'].append(user['lastName']);
    data['headline'].append(user['headline']);
    for activity in activity_group:
        if activity in user:
            data[activity].append(user[activity]);
        else:
            data[activity].append(0);
            
df_users = pd.DataFrame(data, index=data['id']);
# Transform data
df_users['total activity'] = 0;
df_users['total discussion'] = 0;
df_users['total comment'] = 0;
df_users['max activity in a group'] = df_users[activity_group].max(axis=1);
df_users['most active group'] = None;
for activity in activity_group:
    df_users['most active group'][df_users[activity] == df_users['max activity in a group']] = activity;
    df_users['total activity'] += df_users[activity];
    if activity.startswith('discussion'):
        df_users['total discussion'] += df_users[activity];
    else:
        df_users['total comment'] += df_users[activity];
df_users['max activity in a group (%)'] = (df_users['max activity in a group'] * 100.0 / df_users['total activity']).round(2);
# show 10 random users
df_users.ix[np.random.choice(df_users.index, 10)][[
    'firstName',
    'lastName',
    'headline',
    'total activity',
    'total discussion',
    'total comment',
    'most active group',
    'max activity in a group (%)' 
]]
```

# Extract words from Linkedin profiles

```python
textFields = [
    ['headline'],
    ['summary'],
    ['specialties'],
    ['interests'],
    ['publications', -1, 'title'],
    ['publications', -1, 'summary'],
    ['patents', -1, 'title'],
    ['patents', -1, 'summary'],
    ['certifications', "values", -1, 'name'],
    ['positions', 'values', -1, 'title'],
    ['positions', 'values', -1, 'summary'],
    ['positions', 'values', -1, 'company', 'name'],
    ['skills', 'values', -1, 'skill', 'name'], 
    ['educations', 'values', -1, 'school-name'],
    ['educations', 'values', -1, 'field-of-study'],
    ['educations', 'values', -1, 'degree'],
    ['educations', 'values', -1, 'activities'],
    ['educations', 'values', -1, 'notes'],
    ['courses', -1, 'name'],
    ['job-bookmarks'],
    ['recommendations-received', -1, 'recommendation-text']
];

if not os.path.exists("user_wordcloud"):
    os.makedirs("user_wordcloud"); 

c = 0;
usersCounts = {};
for user in userProfiles.values():
    if c < 5:
        print ""
        print ""
        print "User: " + user['id'];
        print ""
        print ""
    text = "";
    for path in textFields:
        #print json.dumps(user, indent=1)
        step = user;
        for i in range(len(path)):
            textField = path[i];  
            if textField == -1:
                arr = step;
                if arr != None and len(arr) > 0:
                    for j in range(len(arr)):
                        k = i + 1;
                        step = arr[j];
                        while k < len(path):
                            textField = path[k];
                            if textField in step:
                                step = step[textField];
                            else:
                                step = None;
                                break;
                            k += 1;
                        if step != None:
                            #print str(path) + "[" + str(j) +"]";
                            #print step;
                            text += " " + step;
                step = None;
                break;
            elif textField in step:
                step = step[textField];
            else:
                step = None;
                break;
        if step != None:
            #print str(path)
            text += " " + step;
    #print text
    tokens = nltk.word_tokenize(text.lower());
    #word_tag_fd = nltk.FreqDist(nltk.pos_tag(tokens));
    word_tag_fd = nltk.FreqDist(tokens);
    if len(text) > 10 and c < 5:
        wordcloud = WordCloud().generate(text.lower())
        plt.imshow(wordcloud)
        plt.axis("off")
        plt.savefig('user_wordcloud/' + user['id'] + '.png')
        plt.show()
        c += 1;
    userCounts = {};
    for (wt, count) in word_tag_fd.most_common():
        #if wt[1].startswith("NN"):
            #userCounts[wt[0]] = count;
            #print wt[0], count
        userCounts[wt] = count; 
        #print wt, count
    usersCounts[user['id']] = userCounts;
f = open("usersCounts.json", 'w');
f.write(json.dumps(usersCounts, indent=1));
f.close();
```

# Generate network of Users' linkedIn profiles

You can open the produced file in UCInet or Gephi for visualizing the network and calculate graph metrics

```python
def ix(x):
    return str(1 + nodeArray.index(x));

def hasCharacters(word):
    result = False;
    for c in word:
        if c.isalpha():
            result = True;
    return result;

nodes = set();
def extract_relation(user):
    char_to_remove = "',{}|<>/;\(),";
    name = user['firstName'] + " " +  user['lastName'];
    nodes.add(name.replace(char_to_remove, ""));
    for word in usersCounts[user['id']]:
        if hasCharacters(word):
            nodes.add(word);
    
active = df_users.sort(['total activity'], ascending=False).head(15);
for userId in active['id']:
    if userId != 'private':
        extract_relation(active.ix[userId]);
nodeArray = list(nodes);

f = open('users.dl','w');
f.write('dl n=' + str(len(nodes)) + '\n');
f.write('format = edgelist1\n');
f.write('labels:\n');
line = "";
for node in nodeArray:
    if (len(line) > 0):
        line += ",";
    line += node;
f.write(line.encode('utf8') + '\n');
f.write('data:\n');
for userId in active['id']:
    if userId != 'private':
        user = active.ix[userId];
        name = user['firstName'] + " " +  user['lastName'];
        for word in usersCounts[user['id']]:
            if hasCharacters(word):
                edge = ix(name) + " " + ix(word) + " " + str(usersCounts[user['id']][word]);
                f.write(edge.encode('utf8') + '\n');
f.close();
```
