The dataset used in this example comes from one of the competitions hosted by Kaggle (Global Energy Forecasting Competition 2012 - Load Forecasting).

In addition, a corresponding temperatures time series is given and participants are also warned that some weeks are missing and thus would need to be backcasted or forecasted.

In the context of this notebook, I will only be focusing on one of the datasets: Energy Load history. 

Moreover, for the purposes of this post, I am not interested in solving the competition itself, but only in demonstrating how the use of some Big Data tools such as Spark could help processing this type of time series data by combining more traditional Statistical Tools (such as Python libraries and R) in a more scalable way.

## Data Loading and cleaning

As mentioned in the instructions of the Kaggle competition, multiple weeks of data are missing from the dataset and one of the tasks for the participants is to fill these missing values up in a smart way.

In this notebook, I won't go into forecasting or backcasting this data but simply propagate the last valid observation forward. Obviously this is certainly not a proper method if we are interested in forecasting accuracy afterward but it is the bare minimum required for successfully fitting a first naive ARIMA model.

```python
import pandas as pd

df = pd.read_csv('Load_history.csv')
def convertCol(x):
    return float(str(x).replace(',', '.'));

for i in df.keys():
    if (i.startswith('h')):
        df[i] = df[i].apply(convertCol)
df = df.fillna(method='ffill')

print len(df), "rows of data (", len(df) / len(df['zone_id'].unique()), "per zone )"
df.head(5)
```

## Data transformation

When looking at the table of data obtained previously after some quick cleaning, we observe that each row corresponds to a day of measurements for each zone at different dates. There are 4 columns describing the location and the date and then 24 other columns for each hour of the day.

A better format for this data would be to flatten one zone into a single row and have all the measurements as columns for all the dates merged together and ordered through time. Next step, the forecasting task would need to consider one row at a time in order to perform a univariate time series analysis.

After transforming this particular dataset, the first data frame with 28 columns and 33000 rows of data is now reshaped into 20 rows (as we have 20 distinct zones) with 39600 columns (39600 measures taken at different times).

```python
import math

f = open('data.csv', 'w');
for zone_id, zone in df.groupby('zone_id'):
    #f.write(str(zone_id));
    #index = ['zone_id']
    index = []
    for row in zone.sort(['year', 'month', 'day']).values:
        basekey = str(int(row[1]));
        for i in range(2, 4):
            basekey += "-" + str(int(row[i]))
        for j in range(4, len(row)):
            key = basekey + ':' + str(j - 4)
            if len(index) > 0:
                f.write('\t')
            index.append(key);
            f.write(str(row[j]));    
    f.write('\n');
f.close();

result = pd.read_csv('data.csv', names=index, sep='\t')
print len(result), "rows of data"
result.head(5)
```

# PySpark environment (ie Python + Spark):

Before going further in describing the time series analysis using ARIMA models, let's first take a closer look at the gain that Spark can provide us in terms of scalability.

In the next section, I will go through a simple computation of the average energy consumption calculation for each zone. Even though this is executed on a single Spark node for testing purposes, it should still illustrate the advantages of using a framework like Spark in terms of speeding up the tasks.

The benchmark test consists of the same function "testParallelization" defined below that will be executed in two scenarios:
- Input data is not partitioned
- Input data is partitioned such as each line of data is in a different partition

In Spark, the data is represented into Resilient Distributed Dataset (RDD) that can be partitioned and distributed across the nodes of the cluster. As a result, in the first scenario, if the RDD is not partitioned, a single Spark node should process the whole data (20 rows) whereas in the second scenario, multiple Spark nodes could potentially be processing each line separately. In the best case scenario, we could have 20 different computers working simultaneously, each one processing a different line (if we had such a cluster size and the data was distributed on the hard-drives such as each partition can be retrieved on distinct machines)

## Spark parallelization test

```python
import numpy as np
import time
import shutil
import datetime

def testParallelization(rdd, delay):
    start = datetime.datetime.now()
    # clean up the folder where to store the output files first
    shutil.rmtree('output', ignore_errors=True)
    # Function to be executed in parallel on each row of data
    def computeMean(row):
        if delay > 0:
            # add a delay to the processing to simulate a long computation
            time.sleep(delay)
        # because we are executing this code on a single node for now,
        # every worker ends up writing in the same file 
        # (similar to multiple threads or process writing to the same file)
        f = open("output/test.log", "a")
        f.write('Processing zone id ' + str(row[0]) + '\n')
        f.close()
        return np.mean(row[1])
    rddres = rdd.map(computeMean)
    rddres.saveAsTextFile("output")
    f = open("output/test.log", "r")
    for line in f:
        print line.strip('\n')
    print "Parallelization test was executed in", (datetime.datetime.now() - start).total_seconds(), "seconds"
```

The execution trace of the first scenario shows that each zone has been processed sequentially and in order (the input was sorted by zone id, and the output came out in the same order). The total execution time took approximatively 3 seconds.

### Using 20 partitions (parallelization to 20 tasks)

```python
rdd = sc.parallelize(zip(result.index, result.values), 20)
testParallelization(rdd, 0.1)
```

The execution trace of the second scenario shows that every partitions has been submitted for processing at the same time. Next they were randomly selected by whatever worker that was available for execution which is why the order of the output is not the same as the input data like in the previous scenario.

The most interesting part is that there is already a significant improvement in overall processing time as it took approximatively one second to run through all the zones. This can be explained even on a single spark node by the fact that it is similar to multi-threading processing where multiple averages were being computed parallel at the same time whereas the first scenario was computing average one line at a time sequentially in a mono-thread manner.

As a conclusion, Spark clusters would allow parallelization to a large number of simultaneous tasks whereas one single computer could process in parallel up to the number of CPUs on the machine only...

# PySpark environment + external process

Instead of using a complex mix of technology to make Spark work with both Python and R, I did actually investigate another approach of directly using <a href="http://amplab-extras.github.io/SparkR-pkg/">SparkR front-end</a> of Spark. Below is the code of my example using the pipe function from pySpark to build an RDD created by piping elements to a forked external process. The RDD is built based on the standard output of my external process where I am printing the coefficients of my ARIMA model for each row of data (each zone).

Note that some plot graphics are also being generated in the output folder for further inspections of the results. and it is working as expected: it is using different parameters for each model of each zone.

```python
import numpy as np
import statsmodels.api as sm

def toString(x):
    s = "";
    for val in x[1]:
        if (len(s) > 0):
            s += '\t';
        s += str(val);
    return str(x[0]) + '\t' + s;

def processInParallel(rdd):
    rddres = rdd.map(toString).pipe('Rscript spark.r output/spark')
    return rddres
   
rdd = sc.parallelize(zip(result.index, result.values), 20)
start = datetime.datetime.now()
for line in processInParallel(rdd).collect():
    print line
print "ARIMA calculations was executed in", (datetime.datetime.now() - start).total_seconds(), "seconds"
```
